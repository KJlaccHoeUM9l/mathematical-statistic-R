# Основные законы распредления.

В базовой версии R реализованы следующие законы распределения вероятностей:
- `*beta` - бета-распределение
- `*binom` - биномиальное распределение (включая распределение Бернулли)
- `*cauchy` - распределение Коши
- `*chisq` - распределение хи-квадрат
- `*exp` - экспоненциальное распределение
- `*f` - распределение Фишера
- `*gamma` - гамма-распределение
- `*geom` - геометрическое распределение
- `*hyper` - гипергеометрическое распределение
- `*lnorm` - логнормальное распределение
- `*multinom` - полиномиальное распределение
- `*nbinom` - отрицательное биномиальное распределение
- `*norm` - нормальное распределение
- `*pois` - распределение Пуассона
- `*t` - распределение Стьюдента
- `*unif` - равномерное распределение
- `*weibull` - распределение Вейбулла

Здесь символ звездочка обозначает то, что именно необходимо от закона распределения:
- `d` (от "density", плотность) - функция плотности вероятности
- `p` (от "probability", вероятность) - интегральная функция распределения вероятностей
- `q` (от "quantile", квантиль) - функция для нахождения квантилей распределения
- `r` (от "random", случайный) - функции для генерации случайных чисел в соответствии с параметрами закона распределения вероятностей

## Основные операции с выборками элементов

### Создание выборки
Создадим выборку из 10 случайных элементов, распределенных по нормальному закону:

```r
X<-rnorm(10)
print(X)
```

Значениями по умолчанию для медианы и среднеквадратичного отклонения является пара (0, 1). Чтобы задать свои параметры, их нужно указать в аргументах:

```r
mean = 4
var = 0.5
X<-rnorm(10, mean, var)
print(X)
```

### Гистограмма
График выборочной плотности распределения называется `гистограммой`. Для ее построения используется следующая функция:

```r
n = 1000
mean = 1
var = 0.5

X<-rnorm(n, mean, var)
```

```r
hist(X)
```

Гистограмма показывает абсолютные частоты. Для того, чтобы произвести `нормировку` нужно добавить в аргументы функции `pr=T`:

```r
hist(X, pr=T)
```

### График плотности распределения

Для того, чтобы узнать как ведет себя график истинной функции распределния можно использовать функцию `curve`.

Для ее использования нужно передать ей в аргументы выражение (для кривой), написанное как функция от x. Чтобы график накладывался на гистограмму необходимо дополнительно указать параметр `add=T`.


```r
hist(X, pr=T, col="gold")
curve(dnorm(x, mean, var), add=T, col="blue", lwd=3)
```

### Границы интервалов гистограммы

Для того, чтобы изменить границы интервалов гистограммы необходимо передать в функцию `hist()` праметре `br=`, который может принимать три различных типа значений:
1. Одно целое число - количество интервалов
2. Массив границ
3. Строковая константа - метод, по которому будут расчитываться границы (Sturges, Scott, FD)

Вызов всех типов параметров представлен ниже:

```r
number_of_intervals = 100
hist(X, col="gold", br=number_of_intervals)
```

```r
interval_boundaries = c(-3, -1, 0, 1, 3)
hist(X, col="gold", br=interval_boundaries)
```

```r
hist(X, col="gold", br="Sturges")
```

```r
hist(X, col="gold", br="Scott")
```

```r
hist(X, col="gold", br="FD")
```
