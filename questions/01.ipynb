## 1. Выборка, выборочное распределение одномерной случайной величины. Построение эмпирической плотности и эмпирической функции распределения в пакете R.

### В `R` выборку можно задать руками через функцию [C](https://www.rdocumentation.org/packages/base/versions/3.6.2/topics/c). Она создает вектор значений. 

```r
c(1,2,3)
```

#### Закон распределения случайной величины может быть задан в виде таблицы, функции распределения либо плотности распределения.

На примере нормального распределения посмотрим как можно задать закон распределения в `R`
Подробнее можноп прочитать [здесь](https://www.rdocumentation.org/packages/stats/versions/3.6.2/topics/Normal) и [здесь](http://seankross.com/notes/dpqr/)
* `dnorm` - задает плотность распределения
* `pnorm` - задает функцию распределения
* `qnorm` - задает выбранный квантиль распределения
* `rnorm` - задает распределения инициализируемое рандомными значениями

Так же можно вместо нормального распределения использовать ряд других популярных распределений:
* `unif` - равномерное
* `t` - t распределение Стьюдента
* `binom` - биномиальное распределение
* `chisq` - хи квадрат распределение

и другие

#### Пример

```r
dnorm(c(1,2), mean = 0, sd = 1)
pnorm(c(1,2), mean = 0, sd = 1)
```

#### Изобразим империческую функцию нормального распределения

```r
# Построим империческую функция распределения
x<-rnorm(100, 1, 0.5)
x.f<-ecdf(x)
curve(x.f)


# Построим поверх истиную функцию распределения
curve(pnorm(x, 1, 0.5), col = "purple", add = T)
```

#### Изобразим империческую плотность нормального распределения

```r
xseq<-seq(-4,4,.01)
densities<-dnorm(xseq, 0,1)

plot(xseq, densities, 
     type="l")
```
