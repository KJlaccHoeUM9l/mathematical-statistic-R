## 5. Метод получения выборки значений случайной величины с помощью генератора псевдослучайных чисел. Метод обратной функции.

Получить случайные значения по какому-либо распределению можно вызывав функция с назанием этого распределения и приставкой `r`.

Аргумент это колличество случайных значений в выборке. 

```r
hist(rnorm(100), pr = T, col='blue')
hist(rpois(100, 5), pr = T, add=T, col='red')
hist(runif(100), pr = T, add=T, col='gold')
```

### Метод обратной функции 

#### т.к. интегральная функция равномерно распределена в отрезке от 0 до 1
#### Тогда если брать случайное число из этого отрезка, то это как бы вероятность значения некоторой (моделируемой) случайной величины

#### И тогда приравнивая это значение к интегральной функции распределения через обратную функцию можно найти само значение этой случайной величины, и оно будет как бы из искомого закона распределения

#### Подробнее можно почитать вот [тут](https://mydocx.ru/2-119450.html)

т.е. хотим найти $x$ имея $F(x)$
1. генерим случайные число r
2. приравниваем $r=F(x)$
3. находим $x=F^{-1}(r)$

```r
x<-runif(1000)
hist(x^2, pr = T, col = "yellow")
curve(0.5/(sqrt(x)),add = T)

#подсчет интеграла int(x^2) от 0 до 1, генерируем выборку равн распр, ищем среднее значение
mean(x^2)
```
